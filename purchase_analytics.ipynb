```python
import pandas as pd
df1 = pd.read_csv('order_products__train.csv',error_bad_lines=False)
df11=pd.read_csv('order_products__prior.csv',error_bad_lines=False)
df_o=pd.concat([df1,df11],axis=0)
df_o.to_csv('order_products.csv',index=True)
df_p=pd.read_csv('products.csv',error_bad_lines=False)
```

```python
df_o.head(5)
df_p.head(5)
df3=pd.merge(df_o,df_p[['product_id','department_id']],on='product_id')
df3.head(100)
```

```python
import numpy as np
length=len(df3.department_id)
df3['times']=pd.Series(np.ones(length))
df3['times']=df3['times'].astype('int')
df3['first_time']=[1 if i==0 else 0  for i in df3.reordered ]
df3.info()
df3.head(10)
assert df3['reordered'].notnull().all()
#data={'department_id':[df3.department_id.unique().sort()],}
```

```python
df_gg=pd.DataFrame(df3.groupby('department_id')['times','first_time'].sum())
df_gg.columns=['number_of_orders','number_of_first_orders']
df_gg['percentage']=df_gg.number_of_first_orders/df_gg.number_of_orders
df_gg
df_gg.to_csv('report.csv',encoding='utf-8',index=True)
```
